# Dual-Axis line plot

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y-%m-%d')
```

```python
# Groupby the count of the trips 
df_group = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = 'merge_flag')
print(df['merge_flag'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
df_temp.columns
```

```python
df_temp.head()
```

```python
df_temp.columns
```

```python
fig,ax = plt.subplots(figsize=(10, 5))
sns.lineplot(data = df_temp['bike_rides_daily'], color = "b")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

ax2 = ax.twinx()
sns.lineplot(data = df_temp['avgTemp'], color="r", ax=ax2)
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips in 2022', fontsize = 18)
plt.show()
```

# Box Plot

```python
# Convert started_at and ended_at to datetime format
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])

```

```python
# Calculate trip duration in minutes
df['trip_duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
data = {'rideable_type': ['electric_bike', 'classic_bike', 'electric_bike', 'classic_bike', 'classic_bike'],
        'trip_duration': [15, 45, 75, 120, 180]}

df = pd.DataFrame(data)

# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='rideable_type', y='trip_duration', data=df)
plt.title('Distribution of Trip Duration by Rideable Type')
plt.xlabel('Type of Bike')
plt.ylabel('Trip Duration (minutes)')

plt.show()
```

## The box plot provides a clear comparison of trip durations between electric bikes and classic bikes. The median trip duration for classic bikes is notably higher, suggesting that users tend to ride classic bikes for longer periods. The interquartile range (IQR) for classic bikes is also wider, indicating greater variability in trip durations compared to electric bikes, which have a more consistent trip duration. The whiskers, which represent the spread of the data within 1.5 times the IQR, show that the overall range of trip durations is larger for classic bikes, with both types of bikes showing no significant outliers. This analysis highlights the different usage patterns for the two types of bikes.

```python
grid = sns.FacetGrid(df, col = "rideable_type")
grid.map(sns.histplot, "trip_duration", bins = 2)
plt.show()
```

## The histogram shows that trips using classic bikes tend to be longer, with the majority of trips lasting between 100 and 150 units of time. In contrast, trips on electric bikes are shorter. This indicates that classic bike users may take longer trips compared to electric bike users.
